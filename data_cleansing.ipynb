```python
import csv
import numpy as np
import pandas as pd
import random

f=open('C:/wk/convenience_store.csv',encoding = ' CP949')# 공공데이터 csv 파일로 저장
data=csv.reader(f) #csv 파일 불러오기

df = pd.DataFrame(data) # data를 dataframe형태로 변환
df = df.rename(columns=df.iloc[0]) # 첫번째 행을 칼럼명으로 지정
df = df.drop(index=0, axis=0) # 칼럼명과 겹치는 첫번째 행 삭제
df = df.dropna(axis=0) # 결치값을 가지는 행 삭제

seven_df=df[df['사업장명'].str.contains('세븐일레븐')] #사업장명이 포켓몬 빵을 파는 곳만 추출하여 저장
mini_df=df[df['사업장명'].str.contains('미니스톱')]
e_df=df[df['사업장명'].str.contains('이마트24')]
gs_df=df[df['사업장명'].str.contains('GS25')]
df=pd.concat([seven_df,mini_df,e_df,gs_df])

result=df[df['영업상태명'].str.contains('폐업')].index # 폐업인 행들을 삭제하는 작업
df.drop(result,inplace=True)

df['WGS84위도'].replace('', 'non', inplace=True) #좌표값이 없는 곳에 non 문자를 넣고 non이 있는 행을 삭제하는 과정
result=df[df['WGS84위도'].str.contains('non')].index
df.drop(result,inplace=True)

ran=range(len(df)) #재고 칼럼 만들고 랜덤 값 입력
inventory_ran_data=[]
count=['재고 없음','1개','2개','3개','4개','5개']

for i in ran:
    r = random.choice(count)
    inventory_ran_data.append(r)
df['inventory']=inventory_ran_data

df.rename(columns = {'사업장명':'store_name'},inplace=True)

df.to_csv("data_cleasing_result.csv",index=False) #정제 완료한 데이터를 다시 csv파일로 저장
```

```python
import pandas as pd
import csv
import requests
import folium
import json
#from PIL import Image
#from folium.features import CustomIcon

url = 'http://ip-api.com/json' #접속된 pc의 ip를 가지고 현재 위치 변환해주는 http 사이트 주소 읽어와서 jsonc 형태로 저장후 위도, 경도만 추출
data = requests.get(url)
res = data.json()
my_lat=res['lat']
my_lon=res['lon']

geo_data=pd.read_csv('C:/Users/alstn_bl71ud5/data_cleasing_result.csv',encoding='cp949',engine='python') #정제한 데이터 불러오기와 geo_data 객체 생성
map_geo=folium.Map(location=[my_lat,my_lon],zoom_start=15) #현재 위치를 첫 시작점으로 설정한 map_geo 객체 생성 

for i,store in geo_data.iterrows(): #정제된 데이터에서 위도, 경도 데이터만 추출하여 folium을 통해 맵에 마커로 표시 map_geo 객체에 추가
    inventory=store['inventory']
    latitude=store['WGS84위도']
    longitude=store['WGS84경도']
    iframe=f"재고: {inventory}<br>위도:{latitude}<br>경도:{longitude}"
    popup = folium.Popup(iframe, max_width='200')
    folium.Marker(location=[store['WGS84위도'],store['WGS84경도']],popup=popup,tooltip=store['store_name'],icon=folium.Icon(color='red',icon='star')).add_to(map_geo)

    
folium.Marker(location=[my_lat,my_lon],tooltip='현재 위치',icon=folium.Icon(color='green',icon='star')).add_to(map_geo) #현재 위치는 초록색으로 마커표시

map_geo.save('map_marker_data_.html')
```
